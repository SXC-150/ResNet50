```python
import tensorflow as tf
import os
import numpy as np
from PIL import Image

folder = '/content/drive/MyDrive/CNN_skin_cancer_Check/Flies_to_check'
model_file = '/content/drive/MyDrive/CNN_skin_cancer_Check/conv2d'

read = lambda imname: np.asarray(Image.open(imname).convert("RGB"))
img_file = [read(os.path.join(folder, filename)) for filename in os.listdir(folder)]
img_array = np.array(img_file, dtype='uint8')

model = tf.keras.models.load_model(model_file)

prediction = model.predict(img_array)
print(prediction)
predict_classes=np.argmax(prediction,axis=1)
print(predict_classes)
```
